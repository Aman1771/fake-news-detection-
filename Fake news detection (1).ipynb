```python
data =data_merge.drop(['title','subject','date'],axis =1)
```

```python
data = data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
data.reset_index(inplace =True)
data.drop(['index'],axis=1,inplace =True)
```

```python
def wordopt(text):
    text =text.lower()
    text =re.sub('\[.*?\]','',text)
    text =re.sub("\\W"," ",text)
    text =re.sub('https?://\s+|www\.\S+','',text)
    text =re.sub('<.*?>+','',text)
    text =re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text =re.sub('\n','',text)
    text =re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
x =data['text']
y =data['class']
```

```python
x_train ,x_test,y_train,y_test =train_test_split(x,y,test_size= 0.25)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorization =TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train)
xv_test =vectorization.transform(x_test) 
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()

LR.fit(xv_train ,y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test,y_test)
```

```python
print (classification_report(y_test ,pred_lr))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred_lr)


print("Confusion Matrix:")
print(cm)

fake_news_accuracy = cm[0][0] / (cm[0][0] + cm[0][1])
true_news_accuracy = cm[1][1] / (cm[1][0] + cm[1][1])

print(f"Fake News Accuracy: {fake_news_accuracy:.2f}")
print(f"True News Accuracy: {true_news_accuracy:.2f}")
```

```python
def output_label(n):
    return "Fake News" if n == 0 else "Not Fake News"

def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)

    predictions = {
        "Logistic Regression": output_label(LR.predict(new_xv_test)[0]),
        "Decision Tree": output_label(DT.predict(new_xv_test)[0]),
        "Gradient Boosting": output_label(GB.predict(new_xv_test)[0]),
        "Random Forest": output_label(RF.predict(new_xv_test)[0]),
    }
    
    return predictions

news = input("Enter news text: ")
result = manual_testing(news)
for model, prediction in result.items():
    print(f"{model}: {prediction}")
```
